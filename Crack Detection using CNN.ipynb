```python
# color accepting in terms of RGB
```

```python
# creating two lists
Numeric_image = []
labels = []
```

```python
for filename in os.listdir(positive_image_path):
  specific_img_path = os.path.join(positive_image_path, filename)
  img = cv2.imread(specific_img_path)  #numeric form of image
  img = cv2.resize(img,(64,64))        #resize the image
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  Numeric_image.append(img)
  labels.append(1)
```

```python
# negative image for same approch
```

```python
negative_imgage_path = "/content/surface-crack-detection/Negative"
for filename in os.listdir(negative_imgage_path):
  specific_img_path = os.path.join(negative_imgage_path, filename)
  img = cv2.imread(specific_img_path)  #numeric form of image
  img = cv2.resize(img,(64,64))        #resize the image
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  Numeric_image.append(img)
  labels.append(0)
```

```python
Numeric_image = np.array(Numeric_image)
labels = np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Numeric_image, labels,random_state = 100, test_size = 0.2 )
```

```python
X_train.shape  #33111 is number of image, here 64 rows, 64 col, 3D data
```

```python
X_test.shape
```

```python
first_image = X_train[1]
```

```python
first_image
```

```python
# normalise the data
first_image = first_image / 255
# 255 is optimal value to normalise our images
# each value between 0 and 1
```

```python
first_image
```

```python
# normalise the data
X_train = X_train / 255
X_test = X_test / 255
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# define the model
model = Sequential()  # brain without neuron
```

```python
model.add(Conv2D(32,(3,3), activation = 'relu', input_shape = (64,64,3) ))  # 32 is kernling with dim 3 by 3
```

```python
model.add(MaxPooling2D(2,2))
```

```python
model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(2,2))
```

```python
model.add(Conv2D(64,(3,3),activation = 'relu'))
```

```python
model.add(Flatten())
```

```python
# create ANN
model.add(Dense(64, activation = 'relu'))
```

```python
# o/p layer
model.add(Dense(1, activation = 'sigmoid'))
```

```python
  # loss and optimiser
model.compile(optimizer = 'adam',loss = "binary_crossentropy", metrics =["accuracy"] )
```

```python
# training the model
model.fit(X_train, y_train ,epochs = 10 , batch_size = 32, validation_data= (X_test, y_test))
```
